```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can replace this with a fine-tuned model if available
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Test the model with a simple prompt
prompt = "Once upon a time"
inputs = tokenizer.encode(prompt, return_tensors="pt")

# Generate text
output = model.generate(inputs, max_length=50, num_return_sequences=1, temperature=0.7)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text:", generated_text)
```

```python
# Generate text with adjusted parameters
output = model.generate(
    inputs,
    max_length=50,
    num_return_sequences=1,
    do_sample=True,  # Enable sampling
    temperature=0.8,  # Adjust randomness
    top_k=50,  # Limit to top-k tokens
    top_p=0.95,  # Use nucleus sampling
    pad_token_id=tokenizer.eos_token_id  # Avoid warning
)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Improved Generated Text:", generated_text)
```

```python
prompt = (
    "In the ancient kingdom of Eldoria, where magic thrived and gods walked among mortals, "
    "a mysterious prophecy spoke of a hero who would rise to restore balance. The story begins with...\n\n"
)

inputs = tokenizer(prompt, return_tensors="pt").input_ids

output = model.generate(
    inputs,
    max_length=150,          # Adjust for length
    do_sample=True,          # Enable sampling
    temperature=1.0,         # Balanced randomness
    top_k=50,                # Top-k sampling
    top_p=0.95,              # Nucleus sampling
    repetition_penalty=1.2,  # Penalize repetition
    pad_token_id=tokenizer.eos_token_id
)

generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text:", generated_text)
```

```python
from nltk.tokenize import sent_tokenize

def clean_repetitions(text):
    sentences = sent_tokenize(text)
    seen = set()
    cleaned = []
    for sentence in sentences:
        if sentence not in seen:
            cleaned.append(sentence)
            seen.add(sentence)
    return " ".join(cleaned)

cleaned_text = clean_repetitions(generated_text)
print("Cleaned Text:", cleaned_text)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load your trained model and tokenizer from the local directory
model = GPT2LMHeadModel.from_pretrained("my_custom_model")
tokenizer = GPT2Tokenizer.from_pretrained("my_custom_model")

# Define the pad_token_id explicitly
model.config.pad_token_id = model.config.eos_token_id  # Set pad_token_id to eos_token_id

# Test generation with a prompt
prompt = "Once upon a time, in a world where"
inputs = tokenizer(prompt, return_tensors="pt")

# Create attention mask (1 for real tokens, 0 for padding tokens)
attention_mask = torch.ones(inputs['input_ids'].shape, dtype=torch.long)

# Generate text using the model
outputs = model.generate(
    inputs['input_ids'],
    max_length=100,
    num_return_sequences=1,
    attention_mask=attention_mask,  # Pass the attention mask here
    no_repeat_ngram_size=2,
    temperature=0.7,
    top_k=50,
    top_p=0.95,
    do_sample=True
)

# Decode and print the generated text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
